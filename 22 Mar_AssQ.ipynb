Q1. Pearson correlation coefficient is a measure of the linear relationship between two variables. Suppose 
you have collected data on the amount of time students spend studying for an exam and their final exam 
scores. Calculate the Pearson correlation coefficient between these two variables and interpret the result.



To calculate the Pearson correlation coefficient, we need to follow these steps:

1. **Collect the data**: Gather pairs of values (study time and exam scores).
2. **Calculate the mean of each variable**.
3. **Subtract the mean from each value**.
4. **Calculate the product of the differences for each pair of values**.
5. **Square the differences for each variable**.
6. **Sum the products and the squared differences**.
7. **Divide the sum of the products by the square root of the product of the sum of squared differences**.

The Pearson correlation coefficient formula is:

\[ $r = \frac{\sum (x_i - \bar{x})(y_i - \bar{y})}{\sqrt{\sum (x_i - \bar{x})^2 \sum (y_i - \bar{y})^2}}$ \]

Where:
- $( x_i )$ and $( y_i )$ are the individual sample points,
- $( \bar{x} )$ and $( \bar{y} )$ are the means of the \( x \) and \( y \) samples, respectively.





Let's go through an example calculation. Suppose we have the following data:

| Study Time (hours) | Exam Score |
|--------------------|------------|
| 2                  | 65         |
| 3                  | 70         |
| 4                  | 75         |
| 5                  | 80         |
| 6                  | 85         |


First, calculate the means of study time $(( \bar{x} ))$ and exam scores $(( \bar{y} ))$:

$[ \bar{x} = \frac{2 + 3 + 4 + 5 + 6}{5} = 4 ]$
$[ \bar{y} = \frac{65 + 70 + 75 + 80 + 85}{5} = 75 ]$

Next, calculate the deviations from the mean, the product of the deviations, and the squared deviations:

|  Study Time (hours)  | Exam Score  | $ ( x_i - \bar{x} )$  |  $ ( y_i - \bar{y} )$  |  $ ( (x_i - \bar{x})(y_i - \bar{y}) )$  |  $ ( (x_i - \bar{x})^2 )$ |  $ ( (y_i - \bar{y})^2 )$  |
|--------------------|------------|---------------------|---------------------|------------------------------------|------------------------|------------------------|
| 2                  | 65         | -2                  | -10                 | 20                                 | 4                      | 100                    |
| 3                  | 70         | -1                  | -5                  | 5                                  | 1                      | 25                     |
| 4                  | 75         | 0                   | 0                   | 0                                  | 0                      | 0                      |
| 5                  | 80         | 1                   | 5                   | 5                                  | 1                      | 25                     |
| 6                  | 85         | 2                   | 10                  | 20                                 | 4                      | 100                    |

Sum these values:

$ [  \sum (x_i -  \bar{x})(y_i - \bar{y}) = 20 + 5 + 0 + 5 + 20 = 50  ] $

$ [  \sum (x_i -  \bar{x})^2 = 4 + 1 + 0 + 1 + 4 = 10  ] $

$ [  \sum (y_i -  \bar{y})^2 = 100 + 25 + 0 + 25 + 100 = 250  ] $

Finally, calculate the Pearson correlation coefficient:

$[ r = \frac{50}{\sqrt{10 \times 250}} = \frac{50}{\sqrt{2500}} = \frac{50}{50} = 1 ]$

### Interpretation:

The Pearson correlation coefficient $( r )$ is 1, which indicates a perfect positive linear relationship between the study time and the exam scores. This means that as the study time increases, the exam scores also increase proportionally.

Q2. Spearman's rank correlation is a measure of the monotonic relationship between two variables. 
Suppose you have collected data on the amount of sleep individuals get each night and their overall job 
satisfaction level on a scale of 1 to 10. Calculate the Spearman's rank correlation between these two 
variables and interpret the result.



To calculate Spearman's rank correlation, we need to follow these steps:

1. **Rank the data**: Assign ranks to the values of each variable.
2. **Calculate the difference between the ranks of each pair of values**.
3. **Square these differences**.
4. **Apply the Spearman rank correlation formula**.

The formula for Spearman's rank correlation coefficient \( \rho \) (rho) is:

$[ \rho = 1 - \frac{6 \sum d_i^2}{n(n^2 - 1)} ]$

where:
-$ ( d_i )$ is the difference between the ranks of each pair of values,
-$ ( n )$ is the number of pairs of values.

Let's go through an example calculation. Suppose we have the following data:

| Sleep (hours) | Job Satisfaction |
|---------------|------------------|
| 5             | 7                |
| 6             | 8                |
| 7             | 9                |
| 8             | 6                |
| 9             | 10               |

### Step 1: Rank the data

| Sleep (hours) | Rank (Sleep) | Job Satisfaction | Rank (Job Satisfaction) |
|---------------|--------------|------------------|-------------------------|
| 5             | 1            | 7                | 3                       |
| 6             | 2            | 8                | 4                       |
| 7             | 3            | 9                | 5                       |
| 8             | 4            | 6                | 2                       |
| 9             | 5            | 10               | 1                       |

### Step 2: Calculate the differences between the ranks

| Sleep (hours) | Rank (Sleep) | Job Satisfaction | Rank (Job Satisfaction) | \( d_i \) | \( d_i^2 \) |
|---------------|--------------|------------------|-------------------------|-----------|-------------|
| 5             | 1            | 7                | 3                       | -2        | 4           |
| 6             | 2            | 8                | 4                       | -2        | 4           |
| 7             | 3            | 9                | 5                       | -2        | 4           |
| 8             | 4            | 6                | 2                       | 2         | 4           |
| 9             | 5            | 10               | 1                       | 4         | 16          |

### Step 3: Sum the squared differences

$[ \sum d_i^2 = 4 + 4 + 4 + 4 + 16 = 32 ]$

### Step 4: Apply the Spearman rank correlation formula

$[ \rho = 1 - \frac{6 \sum d_i^2}{n(n^2 - 1)} ]$ 

Here, $( n = 5 )$:

$[ \rho = 1 - \frac{6 \times 32}{5(5^2 - 1)} = 1 - \frac{192}{5 \times 24} = 1 - \frac{192}{120} = 1 - 1.6 = -0.6 ]$
### Interpretation

The Spearman's rank correlation coefficient \( \rho \) is -0.6, which indicates a moderate negative monotonic relationship between the amount of sleep and job satisfaction. This means that as the amount of sleep increases, job satisfaction tends to decrease.

Q5. A survey was conducted to examine the relationship between age and preference for a particular 
brand of soft drink. The survey results are shown below:


| Age(Years) | Soft drink Preference |
|------------|-----------------------|
|     25     |        Coke           |
|     42     |        Pepsi          |
|     37     |     Mountain dew      |
|     19     |        Coke           |
|     28     |        Peps           |


The data provided consists of categorical preferences for different brands of soft drinks and ages. To examine the relationship between age and soft drink preference, we can use various methods, but Pearson and Spearman correlations are not suitable since they measure relationships between numerical variables.

Instead, we can use Chi-Square test for independence to determine if there is a significant association between age and soft drink preference. Here's the step-by-step process:

1. **Organize the data into a contingency table**.
2. **Calculate the expected frequencies**.
3. **Apply the Chi-Square formula**.

### Step 1: Organize the Data

Given the data:

| Age (Years) | Soft Drink Preference |
|-------------|-----------------------|
| 25          | Coke                  |
| 42          | Pepsi                 |
| 37          | Mountain Dew          |
| 19          | Coke                  |
| 28          | Pepsi                 |

We'll organize this into a contingency table:

| Age Group | Coke | Pepsi | Mountain Dew | Total |
|-----------|------|-------|--------------|-------|
| 19-29     | 2    | 1     | 0            | 3     |
| 30-39     | 0    | 0     | 1            | 1     |
| 40-49     | 0    | 1     | 0            | 1     |
| Total     | 2    | 2     | 1            | 5     |

### Step 2: Calculate the Expected Frequencies

The expected frequency for each cell in the table is calculated using the formula:

$[ E_{ij} = \frac{(row \, total \times column \, total)}{grand \, total} ]$

Where $( E_{ij} )$ is the expected frequency for cell $( (i, j) )$.

| Age Group | Coke | Pepsi | Mountain Dew | Total |
|-----------|------|-------|--------------|-------|
| 19-29     | $(\frac{3 \times 2}{5} = 1.2)$ | $(\frac{3 \times 2}{5} = 1.2)$ | $(\frac{3 \times 1}{5} = 0.6)$ | 3     |
| 30-39     | $(\frac{1 \times 2}{5} = 0.4)$ | $(\frac{1 \times 2}{5} = 0.4)$ | $(\frac{1 \times 1}{5} = 0.2)$ | 1     |
| 40-49     | $(\frac{1 \times 2}{5} = 0.4)$ | $(\frac{1 \times 2}{5} = 0.4)$ | $(\frac{1 \times 1}{5} = 0.2)$ | 1     |
| Total     | 2    | 2     | 1            | 5     |

### Step 3: Apply the Chi-Square Formula

The Chi-Square statistic is calculated using the formula:

$[ \chi^2 = \sum \frac{(O_{ij} - E_{ij})^2}{E_{ij}} ]$

Where $( O_{ij} )$ is the observed frequency, and $( E_{ij} )$ is the expected frequency.

Let's calculate this:

$[
\chi^2 = \frac{(2 - 1.2)^2}{1.2} + \frac{(1 - 1.2)^2}{1.2} + \frac{(0 - 0.6)^2}{0.6} + \frac{(0 - 0.4)^2}{0.4} + \frac{(0 - 0.4)^2}{0.4} + \frac{(1 - 0.2)^2}{0.2} + \frac{(0 - 0.4)^2}{0.4} + \frac{(1 - 0.4)^2}{0.4} + \frac{(0 - 0.2)^2}{0.2}
]$



```python
import numpy as np
from scipy.stats import chi2_contingency

# Create the contingency table
data = np.array([[2, 1, 0], 
                 [0, 0, 1], 
                 [0, 1, 0]])

# Perform the Chi-Square test
chi2, p, dof, expected = chi2_contingency(data)

chi2, p, dof, expected
```
